```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1) ## 布局2行，2列，最多4个子图, 第1张图; 第1行第2列没有图
ax2 = fig.add_subplot(2, 2, 3) ## 在第二行第一个位置的子图；
ax3 = fig.add_subplot(2, 2, 4) ## 第二行第二列的位置的子图；

from numpy.random import randn  # randn---N(0,1) 正态分布，大部分值在 -1.96 - 1.96 之间
import numpy as np

r = randn(50)
plt.plot(r.cumsum(), "k--") ## 会在最后一个用过的子图上面绘制

ax1.hist(randn(100), bins=20, color="k", alpha=0.3)  # 直方图
ax2.scatter(np.arange(30), np.arange(30) + 3 * randn(30))  # 散点图
plt.show()
```

```python
#线形图还可以加上一些标记(marker),以强调实际的数据点
plt.plot(randn(30).cumsum(),color='k',linestyle='dashed',marker='o')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
u = np.linspace(-1,1,100)
x,y = np.meshgrid(u,u)    #meshgrid函数创建一个二维的坐标网络
z = x**2 + y**2
ax.plot_surface(x,y,z,rstride=4,cstride=4,cmap=cm.YlGnBu_r)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm, colors
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# center and radius
center = [1,2,3]
radius = 10

# data
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = radius * np.outer(np.cos(u), np.sin(v)) + center[0]
y = radius * np.outer(np.sin(u), np.sin(v)) + center[1]
z = radius * np.outer(np.ones(np.size(u)), np.cos(v)) + center[2]

# plot
fig = plt.figure()
ax = fig.add_subplot(121, projection='3d')

# surface plot
ax.plot_surface(x, y, z,  rstride=4, cstride=4, color='b')

# wire frame
ax = fig.add_subplot(122, projection='3d')
ax.plot_wireframe(x, y, z, rstride=10, cstride=10)

# show
plt.show()
```
